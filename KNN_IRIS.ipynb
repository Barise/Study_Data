```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()


print("iris_dataset의 키들 : \n{}".format(iris_dataset.keys()))
print("target_names : \n{}".format(iris_dataset['target_names']))
print("featrue_names : \n{}".format(iris_dataset['feature_names']))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    iris_dataset['data'], iris_dataset['target'], random_state=0)
print("X_train 의 크기 : {}".format(X_train.shape))
print("y_train 의 크기 : {}".format(y_train.shape))
print("X_test 의 크기 : {}".format(X_test.shape))
print("y_test 의 크기 : {}".format(y_test.shape))
```

```python
!pip install mglearn

import pandas as pd
import mglearn
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

pd.plotting.scatter_matrix(iris_dataframe, c=y_train, figsize=(12,12), marker='o',
                           hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train, y_train)
```

```python
import numpy as np
X_new = np.array([[7,2.1,6.9,0.3]])

prediction = knn.predict(X_new)

print("예측한 타깃의 이름 : {}".format(iris_dataset['target_names'][prediction]))
```

```python
y_prediction = knn.predict(X_test)
#X_test(data)에 대하여 품종예측

for i in range(0,len(y_prediction)):
    y_pred = y_prediction[i]
    print("{} : {}".format(X_test[i],iris_dataset['target_names'][y_pred]))
```

```python
print("테스트 세트의 정확도 :",knn.score(X_test, y_test))
```

```python
import pandas as pd

row_data = pd.read_csv('IRIS.csv')
row_data.head()
```

```python
## 산점도
import matplotlib.pyplot as plt

plt.scatter(setosa_sepal_length, setosa_sepal_width, label='setosa')
plt.scatter(versicolor_sepal_length, versicolor_sepal_width, label='versicolor')
plt.scatter(virginica_sepal_length, virginica_sepal_width, label='virginica')

plt.xlabel('s_length')
plt.ylabel('s_width')
plt.legend()
plt.show()

##대체로 lenth가
##setosa > versicolor > virginica
```

```python
#데이터 합치기

iris_sepal_length = setosa_sepal_length + versicolor_sepal_length + virginica_sepal_length
iris_sepal_width = setosa_sepal_width + versicolor_sepal_width + virginica_sepal_width

iris_data = np.column_stack((iris_sepal_length, iris_sepal_width))
```

```python
iris_target = np.concatenate((np.full(len(setosa_sepal_length),2), np.ones(len(versicolor_sepal_length)), np.zeros(len(virginica_sepal_length))))
```

```python
#훈련 데이터 / 테스트 세트 나누기
from sklearn.model_selection import train_test_split

train_input, test_input, train_target, test_target = train_test_split(
    iris_data, iris_target, stratify=iris_target
)
```

```python
#표준점수로 데이터 표준화

mean = np.mean(train_input, axis=0)
std = np.std(train_input, axis=0)

train_scaled = (train_input - mean) / std
```

```python
#이상치 제거 (abs 2)
outline = []
for i, sample in enumerate(train_scaled):
    for n in sample:
        if abs(n) >= 2:
            outline.append(i)
            break

print(train_scaled[outline])

train_input = np.delete(train_input, outline, axis=0)
train_target = np.delete(train_target, outline, axis=0)


```

```python
#KNN 알고리즘 분류
from sklearn.neighbors import KNeighborsClassifier

kn = KNeighborsClassifier()

kn.fit(train_input, train_target)
kn.score(test_input, test_target)
```

# 성능 높이기


```python
import pandas as pd

iris = pd.read_csv('IRIS.csv')

iris.head()

```

```python
iris['id'] = range(len(iris))

iris = iris[['id','sepal_length','sepal_width','petal_length','petal_width','species']]

iris.head()
```

```python
#데이터 개수

iris['species'].value_counts()
```

```python
# 데이터 100개만 랜덤으로 추출해서 학습데이터셋

train = iris.sample(100, replace=False, random_state=7).reset_index(drop=True)
train
```

```python
# 학습 데이터에 없는거는 테스트 데이터

test = iris.loc[ ~iris['id'].isin(train['id']) ]
test = test.reset_index(drop=True)
```

```python
#KNN 사용

from sklearn.neighbors import KNeighborsClassifier

#3으로 설정
knn = KNeighborsClassifier(n_neighbors=3)

knn.fit( train[['sepal_length','sepal_width','petal_length','petal_width']] , train['species'] )

predictions = knn.predict( test[['sepal_length','sepal_width','petal_length','petal_width']] )

test['pred'] = predictions

test.head()
```

```python
(test['pred'] == test['species']).mean()
```

```python
# 최적의 KNN의 k값 찾기

for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit( train[['sepal_length','sepal_width','petal_length','petal_width']] , train['species'] )
    predictions = knn.predict( test[['sepal_length','sepal_width','petal_length','petal_width']] )
    print((pd.Series(predictions) == test['species']).mean())


# 최적은 0.98 (K=5)
```

```python
#cross validation 사용 => 데이터가 바뀔떄마다 값이 달라질수 있으니, 교차검증 사용

from sklearn.model_selection import cross_val_score
import numpy as np

for k in range(1,30):
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, iris[['sepal_length','sepal_width','petal_length','petal_width']], iris['species'], cv=5)
    print(f"{k} : " ,np.mean(scores))

# 0.9800000000000001 일때가 가장 높음(K=6)

# 참조 블로그 https://silvercoding.tistory.com/64
```
